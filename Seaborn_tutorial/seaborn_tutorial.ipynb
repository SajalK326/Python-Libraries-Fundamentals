# Relational Plot

- used to see the statistical relationship between two variables.
- Bivariate Analysis

Plots under this category:
- Scatter Plots
- Line Plots

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
# Scatter Plot -> Using Axes level function sns.scatterplot()
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='time', size='size')

# hue parameter -> used to differentiate data using different colors
# style parameter -> used to differentiate data using different styles and shapes
# size parameter -> used to differentiate data using different sizes
```

```python
# Scatter Plot -> Using Figure level function sns.relplot()
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', hue='sex', style='time', size='size')
```

```python
gap=px.data.gapminder()
gap.head()
```

```python
# filtering out the rows for India
gap_india=gap[gap['country']=='India']
gap_india
```

```python
# Axes level function
sns.lineplot(data=gap_india, x='year', y='lifeExp')
```

```python
# Figure level function
sns.relplot(data=gap_india, x='year', y='lifeExp', kind='line')
```

```python
gap_icp=gap[gap['country'].isin(['India', 'Brazil', 'Germany'])]
gap_icp.sample(5)
```

```python
# Multiple plots in a lineplot
sns.relplot(data=gap_icp, x='year', y='lifeExp', kind='line', hue='country', style='continent')

# Here, the plot shows changes in life expectancy over the years of India, Brazil and Germany
# The plots are differentiated according to the country, shown using different colours
# They are also differentiated according to the continent, shown using different styles
```

```python
# Facet Plots -> Only works with Figure level functions
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', hue='sex')

# This is a simple scatter plot
```

```python
# Now, if we want to display the plots of males and females separately
# we can do so by replacing hue parameter with col parameter -> for plotting in different columns
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', col='sex')
```

```python
# We can also use row parameter -> for plotting in different row
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', row='sex')
```

```python
# We can also use both if we want to display more info
# this will create a grid of plots
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', row='sex', col='smoker')
```

```python
# Now if we have a greater number of plots, we can use col_wrap parameter to wrap around the plots
# it take a number as input which represents the number of plots in a column
sns.relplot(data=gap, x='lifeExp', y='gdpPercap', kind='scatter', col='year', col_wrap=4)
```

# Distribution Plots

- used for Univariate Analysis
- used to find out the distribution
- shows the range of the observation
- used to find out the Central Tendency
- find out id the data is bimodal? (If the data has more than one peaks)
- used to find out any outliers in the data

Plots under this category:
- Histogram Plot
- KDE Plot
- Rug Plot

<h3>Histogram

```python
tips.head()
```

```python
# Axes Level function -> histplot
sns.histplot(data=tips, x='total_bill')
```

```python
# Figure Level function -> displot
sns.displot(data=tips, x='total_bill', kind='hist')

# bins parameter -> used to set the no. of bins to divide the data into
```

```python
# hue parameter -> can be used to differentiate the data using different colours
sns.displot(data=tips, x='tip', kind='hist', hue='sex')
```

```python
# for more clarity, there is an option to make the plot translucent, so the other plots are visible
sns.displot(data=tips, x='tip', kind='hist', hue='sex', element='step')
```

```python
# Apparently, the histogram can also be plot on categorical columns
# Then it is not basically a histogram, it is somewhat like a countplot
sns.histplot(data=tips, x='day')
```

<h4>faceting can also be done on these type of plots, using the same methods as before, but here also, it will only work on Figure level functions

<h3> KDE Plot

- Stands for Kernel Density Estimation
- We do not use bins here, instead, it uses a Gaussian Kernel to smoothen the observations, which produces a continuous density estimate

```python
sns.kdeplot(data=tips, x='total_bill')

# the same parameters as histplot can be used for kdeplot (like hue, element, etc.)
# KDE Plot can also be plotted using Figure level function (displot)
# It just needs an extra parameter, i.e., kind='kde'
```

```python
# fill parameter -> it gives a filled KDE Plot
sns.kdeplot(data=tips, x='total_bill', hue='sex', fill=True)
```

```python
# Resizing a plot using a figure level function
sns.displot(data=tips, x='total_bill', hue='sex', fill=True, kind='kde', height=5, aspect=2)

# height Parameter -> Adjusts the height of the figure
# aspect Parameter -> Controls the width-to-height ratio of each individual subplot (facet) within the figure
# Here, in this figure, we have used aspect=2, so it makes the width of the figure double to its height
```

<h3>Rugplot

- It is used to plot marginal distributions by drawing ticks along the x and y axes
- it is intended to complement other plots by showing the location of individual observations

```python
sns.kdeplot(data=tips, x='total_bill')
sns.rugplot(data=tips, x='total_bill')

# It can also be divided on the basis of hue
```

<h3> Bivariate Histogram </h3>

- A bivariate histogram creates bins that represent rectangle pieces
- The count of observations is represented by the colour of each rectangle

```python
sns.histplot(data=tips, x='total_bill', y='tip')

# Similar to a Heatmap
# This can also be done using displot
```

<h3> Bivariate KDE Plot</h3>

- It uses a 2D Gaussian Function to smoothen the (x,y) observations

```python
sns.kdeplot(data=tips, x='total_bill', y='tip')

# Similar to a Contour Plot
# Greater density of lines -> more customers
# Lesser density of lines -> less customers

# This can also be done using displot
```

# Matrix Plots

Plots under this category:
- Heatmap
- Clustermap

<h3>Heatmap -> Plots data as a rectangular colour coded matrix

```python
gap.head()
```

```python
# The above data is in 'long data' format, we will convert this into 'wide data' format
# We will do so by making a pivot table with Columns as Years, Rows as Country and values as lifeExp

temp_df=gap.pivot(index='country', columns='year', values='lifeExp')
temp_df.head()
```

```python
# Heatmap using axes level function
plt.figure(figsize=(15,12))     # Adjusting the size of the figure
sns.heatmap(data=temp_df)
```

<h4> <b>Note:</b> Heatmap does not have any Figure level function

```python
# Lets create a heatmap which contains the lifeExp of only the countries of Europe

temp_df2=gap[gap['continent']=='Europe'].pivot(index='country', columns='year', values='lifeExp')
temp_df2.head()

# Here, first, we filter out all the european countries, then make a similar pivot table out of them as we made before
```

```python
# Now, heatmap creation
plt.figure(figsize=(12,8))
sns.heatmap(data=temp_df2, annot=True, linewidth=0.5, cmap='viridis')

# annot parameter -> (Boolean) This is used to display the values in of the observations in the heatmap
# linewidth parameter -> This is used to give spaces between boxes in the graph for better readability
# cmap parameter -> This is used to change the colour scheme of the graph
```

<h3> Clustermap

- This plots a Hierarchically Clustered Heatmap

<b>Note:</b> It requires scipy to be available

```python
# Lets load the iris dataset from plotly
iris=px.data.iris()
iris.sample(5)
```

```python
# Creating the Clustermap
sns.clustermap(iris.iloc[:, [0,1,2,3]])

# We passed the columns for which we want to make a clustermap
# : -> passes all the rows
# [0,1,2,3] -> selected the first 4 columns of the dataset
```

<h3> Strip Plot </h3>

Used for plotting a scatter plot between a categorical and a numerical column

```python
# Axes level function
sns.stripplot(data=tips, x='day', y='total_bill')

# jitter parameter -> (boolean/float) It is used to add/remove extra noise, which is added by default by seaborn internally to make the plot more readable.
# All the points in a particular day should have been in a straight line, but this noise helps to make the graph more readable.

# hue parameter -> used to differentiate the data using colours
```

```python
# figure level function -> catplot()
sns.catplot(data=tips, x='day', y='total_bill', kind='strip')
```

<h3>Swarm Plot

It runs an algorithm internally which arranges data in a swarm like pattern, which somewhat represents the original data distribution

It is better for small data

```python
sns.swarmplot(data=tips, x='day', y='total_bill')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='swarm')
```

<h3>Boxplots

- It plots the distribution of data based on a five-number summary.
- Namely: Minimum, Maximum, 1st Quartile, 3rd Quartile and the Median.
- It can tell if the data is symmetrical, the tightness of grouping of the data, and if the data is skewed.
- It can also give details about the outliers in the data.

```python
# Axes level function
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex')
```

```python
# Figure level function
sns.catplot(data=tips, x='day', y='total_bill', kind='box', hue='sex')
```

```python
# Single Boxplot
sns.boxplot(data=tips, y='total_bill')
```

<h3>Violin Plot

It is a combination of box plot and KDE plot

```python
sns.violinplot(data=tips, x='day', y='total_bill')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='violin', hue='sex', split=True)

# hue parameter -> used to differentiate data using color
# split parameter -> (boolean) used to display two different hue data in a single plot
```

<h3>Barplot

```python
sns.barplot(data=tips, x='sex', y='total_bill', hue='smoker')
```

- The black line is called the error bar. Error bars are used to visualize uncertainty or variability around the estimated value (typically the mean) of each bar.
- By default, Seaborn calculates a 95% bootstrapped confidence interval. This indicates the range in which the true population mean is likely to fall.
- The errorbar can be removed by setting parameter: errorbar=None

```python
# estimator parameter -> This is used to set the estimator fot the barplot.
# like mean, meadian, mode, max, min, standard deviation, etc.
# By default, the estimator is 'mean'.

import numpy as np
sns.barplot(data=tips, x='sex', y='total_bill', hue='smoker', estimator=np.std, errorbar=None)
```

<h3>Point Plot

Similar to barplot, just the difference is that it does not display the barplots.<br>
It displays just the error bars, and the estimator points, and connects them throughout different categories.

```python
sns.pointplot(data=tips, x='sex', y='total_bill')
```

```python
sns.pointplot(data=tips, x='sex', y='total_bill', hue='smoker')

# By default, the estimator is mean.
```

<h3>Countplot

- It is a special case of bar plot.
- It is used if we need to display the no. of observatinos in each category rather than computing a statistic for another variable.
- It is similar to a histogram.

```python
sns.countplot(data=tips, x='sex')

# This plot displays the no. of Males and Females that have visited the place.
```

```python
sns.countplot(data=tips, x='sex', hue='day')

# This displays the day-wise no. of Males and Females that have visited the place.
```

Here also Plot Faceting can be done using catplot()

# Regression Plots

Plots under this category:
- Regplot
- lmplot

Both of the functions:
- plot a scatterplot of two variables x and y
- then fit the regression model y ~ x
- and then plot the resulting regression line and a 95% confidence interval for that regression.

```python
# Axes level function
sns.regplot(data=tips, x='total_bill', y='tip')

# regplot() does not have any hue parameter
```

```python
# figure level function
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex')

# hue parameter is only available in lmplot()
```

<h3>Residual Plot

- It represents the distance of every point from the Best Fit Line of Regression. (Error in prediction)
- 0 represents No error in prediction
- Above 0 -> Positive error in prediction
- Below 0 -> Negative error in prediction

```python
sns.residplot(data=tips, x='total_bill', y='tip')
```

# Plotting Pairwise Relationship Using Multiplots

```python
iris.sample(5)
```

<h3> Pairplot

```python
sns.pairplot(data=iris, hue='species')

# It plots a grid which contains the pairwise relationships in the dataset. 
# It plots a scatter plot for each pair of columns in the dataset. 
# It also plots a histogram for each column on the diagonal (as it is the same column).
# The grid is symmetrical along the diagonal, so the plot in the 1st row and 2nd column is same as the plot in the 2nd row and 1st column, and so on.

# Note: When hue parameter is specified, the diagonal plots are KDE plots instead of histograms, and the scatter plots are coloured according to the hue variable.
```

<h3> Pair Grid

```python
grid2=sns.PairGrid(data=iris, hue='species')
grid2.map(sns.scatterplot)

# The above code is same as the pairplot code, but here we have used PairGrid instead of pairplot.
```

```python
# PairGrid is more flexible than pairplot, as it allows us to use different types of plots for different pairs of columns, while pairplot uses the same type of plot for all pairs of columns.
grid3=sns.PairGrid(data=iris, hue='species')
grid3.map_diag(sns.violinplot)
grid3.map_offdiag(sns.histplot)

# pairplot is a simpler version of PairGrid, which uses the same type of plot for all pairs of columns, while PairGrid allows us to use different types of plots for different pairs of columns.
```

```python
# We can also plot different types of plots for upper and lower triangles of the grid using map_upper() and map_lower() functions respectively.
grid4=sns.PairGrid(data=iris, hue='species')
grid4.map_diag(sns.violinplot)
grid4.map_upper(sns.histplot)
grid4.map_lower(sns.scatterplot)
```

```python
# We can also restrict the plotting to only a few columns of the dataset by passing the subset of columns to the PairGrid function using the `vars` parameter.
grid5=sns.PairGrid(data=iris, hue='species', vars=['sepal_length', 'sepal_width'])
grid5.map_diag(sns.violinplot)
grid5.map_upper(sns.scatterplot)
grid5.map_lower(sns.kdeplot)
```

<h3>Jointplot

```python
sns.jointplot(data=iris, x='sepal_length', y='sepal_width', hue='species', kind='scatter')

# It plots bivariate plots between two variables, along with the univariate plots for each variable on the axes.
# The kind parameter is used to specify the type of plot to be plotted. It can take values like: 'scatter', 'hist', 'hex', 'kde', 'reg', 'resid'
# By default, the kind parameter is 'scatter', so if we do not specify it, it will plot a scatter plot.
```

<h3>Joint Grid

```python
# JointGrid is more flexible than jointplot, as it allows us to use different types of plots for the bivariate and univariate plots, while jointplot uses the same type of plot for both.
grid6=sns.JointGrid(data=iris, x='sepal_length', y='sepal_width', hue='species')
grid6.plot(sns.kdeplot, sns.violinplot)   # Here, the first parameter is for the bivariate plot and the second parameter is for the univariate plot.
```
