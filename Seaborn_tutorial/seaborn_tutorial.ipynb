# Types of functions in Seaborn

<h3>Levels:</h3>

- Figure Level Functions
- Axis Level Functions

<h3>Main Classification:</h3>

- Relational Plot
- Distribution Plot
- Categorical Plot
- Regression Plot
- Matrix Plot
- Multiplots

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

# Relational Plot

- used to see the statistical relationship between two variables.
- Bivariate Analysis

Plots under this category:
- Scatter Plots
- Line Plots

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
# Scatter Plot -> Using Axes level function sns.scatterplot()
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='time', size='size')

# hue parameter -> used to differentiate data using different colors
# style parameter -> used to differentiate data using different styles and shapes
# size parameter -> used to differentiate data using different sizes
```

```python
# Scatter Plot -> Using Figure level function sns.relplot()
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', hue='sex', style='time', size='size')
```

```python
gap=px.data.gapminder()
gap.head()
```

```python
# filtering out the rows for India
gap_india=gap[gap['country']=='India']
gap_india
```

```python
# Axes level function
sns.lineplot(data=gap_india, x='year', y='lifeExp')
```

```python
# Figure level function
sns.relplot(data=gap_india, x='year', y='lifeExp', kind='line')
```

```python
gap_icp=gap[gap['country'].isin(['India', 'Brazil', 'Germany'])]
gap_icp.sample(5)
```

```python
# Multiple plots in a lineplot
sns.relplot(data=gap_icp, x='year', y='lifeExp', kind='line', hue='country', style='continent')

# Here, the plot shows changes in life expectancy over the years of India, Brazil and Germany
# The plots are differentiated according to the country, shown using different colours
# They are also differentiated according to the continent, shown using different styles
```

```python
# Facet Plots -> Only works with Figure level functions
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', hue='sex')

# This is a simple scatter plot
```

```python
# Now, if we want to display the plots of males and females separately
# we can do so by replacing hue parameter with col parameter -> for plotting in different columns
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', col='sex')
```

```python
# We can also use row parameter -> for plotting in different row
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', row='sex')
```

```python
# We can also use both if we want to display more info
# this will create a grid of plots
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', row='sex', col='smoker')
```

```python
# Now if we have a greater number of plots, we can use col_wrap parameter to wrap around the plots
# it take a number as input which represents the number of plots in a column
sns.relplot(data=gap, x='lifeExp', y='gdpPercap', kind='scatter', col='year', col_wrap=4)
```

# Distribution Plots

- used for Univariate Analysis
- used to find out the distribution
- shows the range of the observation
- used to find out the Central Tendency
- find out id the data is bimodal? (If the data has more than one peaks)
- used to find out any outliers in the data

Plots under this category:
- Histogram Plot
- KDE Plot
- Rug Plot

<h3>Histogram

```python
tips.head()
```

```python
# Axes Level function -> histplot
sns.histplot(data=tips, x='total_bill')
```

```python
# Figure Level function -> displot
sns.displot(data=tips, x='total_bill', kind='hist')

# bins parameter -> used to set the no. of bins to divide the data into
```

```python
# hue parameter -> can be used to differentiate the data using different colours
sns.displot(data=tips, x='tip', kind='hist', hue='sex')
```

```python
# for more clarity, there is an option to make the plot translucent, so the other plots are visible
sns.displot(data=tips, x='tip', kind='hist', hue='sex', element='step')
```

```python
# Apparently, the histogram can also be plot on categorical columns
# Then it is not basically a histogram, it is somewhat like a countplot
sns.histplot(data=tips, x='day')
```

<h4>faceting can also be done on these type of plots, using the same methods as before, but here also, it will only work on Figure level functions

<h3> KDE Plot

- Stands for Kernel Density Estimation
- We do not use bins here, instead, it uses a Gaussian Kernel to smoothen the observations, which produces a continuous density estimate

```python
sns.kdeplot(data=tips, x='total_bill')

# the same parameters as histplot can be used for kdeplot (like hue, element, etc.)
# KDE Plot can also be plotted using Figure level function (displot)
# It just needs an extra parameter, i.e., kind='kde'
```

```python
# fill parameter -> it gives a filled KDE Plot
sns.kdeplot(data=tips, x='total_bill', hue='sex', fill=True)
```

```python
# Resizing a plot using a figure level function
sns.displot(data=tips, x='total_bill', hue='sex', fill=True, kind='kde', height=5, aspect=2)

# height Parameter -> Adjusts the height of the figure
# aspect Parameter -> Controls the width-to-height ratio of each individual subplot (facet) within the figure
# Here, in this figure, we have used aspect=2, so it makes the width of the figure double to its height
```

<h3>Rugplot

- It is used to plot marginal distributions by drawing ticks along the x and y axes
- it is intended to complement other plots by showing the location of individual observations

```python
sns.kdeplot(data=tips, x='total_bill')
sns.rugplot(data=tips, x='total_bill')

# It can also be divided on the basis of hue
```

<h3> Bivariate Histogram </h3>

- A bivariate histogram creates bins that represent rectangle pieces
- The count of observations is represented by the colour of each rectangle

```python
sns.histplot(data=tips, x='total_bill', y='tip')

# Similar to a Heatmap
# This can also be done using displot
```

<h3> Bivariate KDE Plot</h3>

- It uses a 2D Gaussian Function to smoothen the (x,y) observations

```python
sns.kdeplot(data=tips, x='total_bill', y='tip')

# Similar to a Contour Plot
# Greater density of lines -> more customers
# Lesser density of lines -> less customers

# This can also be done using displot
```

# Matrix Plots

Plots under this category:
- Heatmap
- Clustermap

<h3>Heatmap -> Plots data as a rectangular colour coded matrix

```python
gap.head()
```

```python
# The above data is in 'long data' format, we will convert this into 'wide data' format
# We will do so by making a pivot table with Columns as Years, Rows as Country and values as lifeExp

temp_df=gap.pivot(index='country', columns='year', values='lifeExp')
temp_df.head()
```

```python
# Heatmap using axes level function
plt.figure(figsize=(15,12))     # Adjusting the size of the figure
sns.heatmap(data=temp_df)
```

<h4> <b>Note:</b> Heatmap does not have any Figure level function

```python
# Lets create a heatmap which contains the lifeExp of only the countries of Europe

temp_df2=gap[gap['continent']=='Europe'].pivot(index='country', columns='year', values='lifeExp')
temp_df2.head()

# Here, first, we filter out all the european countries, then make a similar pivot table out of them as we made before
```

```python
# Now, heatmap creation
plt.figure(figsize=(12,8))
sns.heatmap(data=temp_df2, annot=True, linewidth=0.5, cmap='viridis')

# annot parameter -> (Boolean) This is used to display the values in of the observations in the heatmap
# linewidth parameter -> This is used to give spaces between boxes in the graph for better readability
# cmap parameter -> This is used to change the colour scheme of the graph
```

<h3> Clustermap

- This plots a Hierarchically Clustered Heatmap

<b>Note:</b> It requires scipy to be available

```python
# Lets load the iris dataset from plotly
iris=px.data.iris()
iris.sample(5)
```

```python
# Creating the Clustermap
sns.clustermap(iris.iloc[:, [0,1,2,3]])

# We passed the columns for which we want to make a clustermap
# : -> passes all the rows
# [0,1,2,3] -> selected the first 4 columns of the dataset
```
