# MatPlotLib Advanced

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('iris.csv')
df.sample(5)
```

```python
# Encoding the 'Species' column
df['Species']=df['Species'].replace({'Iris-setosa':0, 'Iris-versicolor': 1, 'Iris-virginica': 2})
df.sample(5)
```

```python
# Plotting the Scatter Plot

plt.figure(figsize=(10,7))      # Used to set the size of the figure
plt.scatter(df['SepalLengthCm'], df['PetalLengthCm'], c=df['Species'], alpha=0.5)
# alpha parameter is used to set the opacity of the graph
# alpha = 1 -> Highest opacity
# alpha = 2 -> Least opacity

plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.colorbar()      # Shows a Colorbar beside the graph
plt.grid()
plt.show()
```

```python
batters=pd.read_csv('batter.csv')
batters.head()
```

```python
batters.shape
```

```python
# Randomly selected 25 batters from Top 100 batters of the dataset
sample_df=batters.head(100).sample(25, random_state=5)
sample_df
```

```python
# Plotting the Scatter Plot
plt.scatter(sample_df['avg'], sample_df['strike_rate'])
plt.xlabel('Average')
plt.ylabel('Strike Rate')
plt.grid()
plt.show()
```

```python
# Annotations -> used for adding annotations/text to the graph
plt.figure(figsize=(10,8))
plt.scatter(sample_df['avg'], sample_df['strike_rate'])

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i], fontdict={'size': 12, 'color': 'orange'})
    # plt.text() is used to add text to the graph
    # it takes (x,y,"text") as parameters
    # fontdict parameter is used to set the font size, color, etc.

plt.xlabel('Average')
plt.ylabel('Strike Rate')
plt.grid()
plt.show()
```

```python
# Drawing Horizontal and Vertical lines in a graph
plt.figure(figsize=(10,8))
plt.scatter(sample_df['avg'], sample_df['strike_rate'])

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i], fontdict={'size': 12, 'color': 'orange'})

plt.axhline(140, color='red')       # Horizontal line -> parameters: (y,color)
plt.axvline(28, color='green')      # Vertical line -> parameters: (x,color)
plt.xlabel('Average')
plt.ylabel('Strike Rate')
plt.grid()
plt.show()
```

```python
# 2-D Subplots

fig, ax=plt.subplots(nrows=2, ncols=2, figsize=(12,12))

ax[0,0].scatter(batters['avg'], batters['strike_rate'])
ax[0,1].scatter(batters['avg'], batters['runs'])
ax[1,0].hist(batters['avg'])
ax[1,1].hist(batters['runs'])

plt.show()
```

```python
# Another way to make 2-D subplots

fig= plt.figure()

ax1=fig.add_subplot(2,2,1)
ax1.scatter(batters['avg'], batters['strike_rate'])

ax2=fig.add_subplot(2,2,2)
ax2.scatter(batters['avg'], batters['runs'])

ax3=fig.add_subplot(2,2,3)
ax3.hist(batters['avg'])

ax3=fig.add_subplot(2,2,4)
ax3.hist(batters['runs'])

plt.show()

# add_subplots() function -> used to add subplots to the figure
# parameters: (nrows, ncols, graph_no)
```

<h3>3-D Surface Plots

```python
x=np.linspace(-10, 10, 100)
y=np.linspace(-10, 10, 100)

xx, yy = np.meshgrid(x,y)   # makes a meshgrid out of x and y of size 100x100
```

```python
z = xx**2 + yy**2
z.shape

# z is also a meshgrid os size 100x100
```

```python
fig=plt.figure(figsize=(12,8))
ax=plt.subplot(projection='3d')

ax.plot_surface(xx, yy, z, cmap='viridis')
plt.show()

# for displaying the colorbar, store the result of plot_surface() in a variable and pass it to plt.colorbar() function
```

<h3>Contour Plot

```python
# Hollow Contour Plot

fig=plt.figure(figsize=(12,8))
ax=plt.subplot()

plot=ax.contour(xx, yy, z, cmap='viridis')
plt.colorbar(plot)
plt.show()
```

```python
# Filled Contour Plot

fig=plt.figure(figsize=(12,8))
ax=plt.subplot()

plot=ax.contourf(xx, yy, z, cmap='viridis')
plt.colorbar(plot)
plt.show()
```

<h3> Heatmap

```python
# We have to make a heatmap of how many sixes were hit on each ball of each over

delivery=pd.read_csv('IPL_Ball_by_Ball_2008_2022.csv')
delivery.head()
```

```python
# By observing the dataset, we get to know that the column 'ballnumber' has more than the values required.
# i.e., instead of 6 balls in every over, there are more than 6 also
# this shows that there were extra balls played (like No Ball, Wide, etc.)

delivery['ballnumber'].unique()
```

```python
# Thus, we will filter out the required data
# We will keep only 6 balls, removing the extras
# We will also keep only those balls in which sixes were hit

data=delivery[(delivery['ballnumber'].isin([1,2,3,4,5,6]) & (delivery['batsman_run']==6))]
data
```

```python
# Now we will create a pivot table and store it in a variable

grid=data.pivot_table(index='overs', columns='ballnumber', values='batsman_run', aggfunc='count')
grid
```

```python
# Now, we will use the imshow() function and pass the grid to plot the heatmap

plt.figure(figsize=(12,8))
plt.imshow(grid)
plt.colorbar()
plt.show()
```
